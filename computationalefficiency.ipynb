```python
import os
import time
import pandas as pd
from kaggle.api.kaggle_api_extended import KaggleApi

# Replace 'some_user/some_dataset' with the actual path of the dataset you want to access
folder_path = r"C:\Users\gbore\Documents\FullDataCsv"

# Start time of the loop
start = time.time()

# Iterate over the files in the folder
dfcompany = []
for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    
    if os.path.isfile(file_path):  # Check if it's a file (not a subdirectory)
        # Perform operations on the file
        df = pd.read_csv(file_path)
        dfcompany.append(df)

#End and total time of the loop
end = time.time()
print("Read csv with pandas: ",(end-start),"sec")
```
